```python
 # Display confusion matrix and classification report for testing set
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_test_pred_lr))

print("\nClassification Report:")
print(classification_report(y_test, y_test_pred_lr))
```

```python
import pandas as pd

# Load your dataset
dataset_path = 'Updated_ Dup_Dataset.csv'
data = pd.read_csv(dataset_path)

# Create a dictionary for symptoms (assuming columns 1 to 378 are the symptoms)
symptoms_dict = {col: idx + 1 for idx, col in enumerate(data.columns[:-1])}


# Print the dictionaries for verification
print("Symptoms Dictionary:", symptoms_dict)
```

```python
# Create a dictionary with unique diseases
diseases_dict = {disease: index for index, disease in enumerate(y.unique())}

print("Diseases Dictionary:",diseases_dict)

# Reverse the diseases_dict
reverse_diseases_dict = {value: key for key, value in diseases_dict.items()}
```

```python
# Reverse dictionary
reverse_diseases_dict = {value: key for key, value in diseases_dict.items()}

def given_predicted_value(user_symptoms):
    # Initialize a zero vector for the input based on the number of symptoms
    input_vector = np.zeros(len(symptoms_dict))
    
    # Iterate over user input symptoms and fill the input vector
    for item in user_symptoms:
        if item in symptoms_dict:
            input_vector[symptoms_dict[item] - 1] = 1  # Mark the symptom as present
        else:
            print(f"Symptom '{item}' not recognized.")
            return None  # Exit if an unrecognized symptom is found
    
    # Make a prediction using the logistic regression model
    prediction = lr_model.predict([input_vector])[0]
    print(f"Model prediction: {prediction}")  # Debugging: print the predicted value

    # Get the disease name from the reverse dictionary
    if prediction in reverse_diseases_dict:
        return reverse_diseases_dict[prediction]
    else:
        print(f"Prediction '{prediction}' not found in the disease dictionary.")
        return "Unknown disease"

# Simulating user input
symptoms = input("Enter your symptoms (comma-separated): ")

# Split the user's input into a list of symptoms (assuming they are comma-separated)
user_symptoms = [s.strip() for s in symptoms.split(',')]

# Get the predicted disease
predicted_disease = given_predicted_value(user_symptoms)

# Print the predicted disease or error message
if predicted_disease:
    print(f"Predicted Disease: {predicted_disease}")
else:
    print("Prediction could not be made. Please check your symptoms.")
```
